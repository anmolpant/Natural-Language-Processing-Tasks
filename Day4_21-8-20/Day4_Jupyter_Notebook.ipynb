```python
# Task 2: Simple Text Classification
def gender_features(word):
    return {'last_letter': word[-1]}
```

```python
gender_features("Obama")
```

```python
from nltk.corpus import names 
```

```python
names.words()
```

```python
print(len(names.words()))
```

```python
labelled_names = ([(name,'male') for name in names.words('male.txt')] + [(name,'female') for name in names.words('female.txt')])
```

```python
import random
random.shuffle(labelled_names)
featuresets = [(gender_features(n),gender) for (n,gender) in labelled_names]
```

```python
train_set,test_set = featuresets[5000:],featuresets[:2000]
import nltk 
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_features('Anmol'))
```

```python
classifier.classify(gender_features('Ayushi'))
```

```python
print(nltk.classify.accuracy(classifier,test_set))
```

```python
# Task 3: Count Vectorizer
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(binary=True)
corpus = ["Walking through the door of this old and lonely place that used to feel like us."]
```

```python
vect.fit(corpus)
```

```python
vocab = vect.vocabulary_
```

```python
for key in sorted(vocab.keys()):
     print("{}:{}".format(key, vocab[key]))
```

```python
print(vect.transform(["This is a good optical illusion"]).toarray())
```

```python
# Finding Similarity Between Documents
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vect.transform(["old and lonely place"]).toarray(), vect.transform(["old place that feels lonely."]).toarray())
print(similarity)
```
